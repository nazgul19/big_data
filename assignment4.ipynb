```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)
```

```python
df.head(5)
```

```python
total_profit = df['total_profit']
```

```python
total_profit
```

```python
plt.plot(df["month_number"],df["total_profit"],color='blue')
plt.xlabel("months")
plt.ylabel("profit")
plt.title("profit for each month")
```

```python
monthlist = df['month_number'].tolist()
facecream = df['facecream'].tolist()
facewash = df['facewash'].tolist()
toothpaste = df['toothpaste'].tolist()
bathingsoap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
moisturizer = df['moisturizer'].tolist()
plt.plot(monthlist,facecream,label = "face cream sales",marker='o',linewidth=2)
plt.plot(monthlist,facewash,label = "face wash sales",marker='o',linewidth=2)
plt.plot(monthlist,toothpaste,label = "toothpaste sales",marker='o',linewidth=2)
plt.plot(monthlist,bathingsoap,label = "bathing soap sales",marker='o',linewidth=2)
plt.plot(monthlist,shampoo,label = "shampoo sales",marker='o',linewidth=2)
plt.plot(monthlist,moisturizer,label = "moisturizer sales",marker='o',linewidth=2)
plt.xlabel('month number')
plt.ylabel('sales units')
plt.title("sale data")
plt.xticks(monthlist)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,14000,16000])
plt.show()
```
